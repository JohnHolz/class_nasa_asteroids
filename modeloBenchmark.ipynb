```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
x,y = get_x_y()
```

```python
def get_simple_metrics(x):
    desc = pd.DataFrame([x.min(),x.max(),x.mean(),x.median(),x.var(),x.kurtosis()]).transpose()
    desc.columns = ['min','max','mean','median','var','kurtosis']
    return desc
```

```python
standard = StandardScaler()
standard.fit_transform(x)
```

```python
get_simple_metrics(x)
```

```python
pd.concat([
    pd.Series(x.columns),
    get_simple_metrics(pd.DataFrame(standard.fit_transform(x)))]         
,axis=1)
```
